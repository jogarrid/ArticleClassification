# Quick data analaysis of dataset extracted


```python
from suds.client import Client
import time
import pandas as pd
import csv
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
nofired_dataset = pd.read_csv('../data/past/UNDISMISSED_final.csv')
fired_dataset = pd.read_csv('../data/past/DISMISSED_final.csv')
```

```python
fired_dataset = fired_dataset.drop_duplicates(subset='author')
nofired_dataset = nofired_dataset.drop_duplicates(subset='author')
```

```python
distr_organisation = fired_dataset.organisation.value_counts() [:10]
distr_organisation.plot.bar(color='b',edgecolor='black')
plt.xlabel('Universities')
plt.ylabel('Occurences')
#plt.title('10 most common affiliations of fired persons')
plt.tight_layout()
plt.savefig('../images/universites_fired.png')
```

```python
U=[]
universities = list(nofired_dataset.organisation)
for unis in universities:
    for u in unis.split(','):
        u = u.replace('[','')
        u = u.replace(']','')
        U.append(u)
```

```python
nofired_unis = pd.DataFrame()
nofired_unis['organisation']=U
```

```python
distr_organisation_un=nofired_unis.organisation.value_counts()[:10]
```

```python
distr_organisation_un.plot.bar(color='b',edgecolor='black')
plt.xlabel('Universities')
plt.ylabel('Occurences')
#plt.title('10 most common affiliations of fired persons')
plt.tight_layout()
plt.savefig('../images/universites_nofired.png')
```
